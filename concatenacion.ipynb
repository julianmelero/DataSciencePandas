```python
import pandas as pd
import numpy as np
```

```python
df_london = pd.read_csv('london_merged.csv')
df_meteorite = pd.read_csv('Meteorite_Landings.csv')
```

# Concatenación de DataSets

```python
x1 = np.random.randn(2,5) * 100
x1
```

```python
x2 = np.random.randn(2,5) * 100
x2
```

```python
np.concatenate([x1,x2])
```

## Eje Columnas

```python
np.concatenate([x1,x2], axis=1).shape
```

## Usando Pandas

```python
s1 = pd.Series(x1[0],index=['a','b','c','d','e'])
s1
```

```python
s2 = pd.Series(x2[0],index=['c','d','e','f','g'])
s2
```

```python
pd.concat([s1,s2])
```

```python
pd.concat([s1,s2], axis=1)
```

```python
s1.reset_index(drop=True)
```

```python
pd.concat([s1.reset_index(drop=True),s2.reset_index(drop=True)], axis=1)
```

## DataFrames

```python
pd.concat([df_london, df_meteorite], axis=1)
```

```python
pd.concat([df_london, df_meteorite], axis=1, join='inner')
```

```python
df_london.append(df_meteorite)
```

```python
df_london.T.append(df_meteorite.T)
```

# Merge
Fuentes de datos diferentes pero se unen por algún campo
